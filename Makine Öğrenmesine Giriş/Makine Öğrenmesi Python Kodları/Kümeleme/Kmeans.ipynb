```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Rastgele veri oluturalm
np.random.seed(42)
n_samples = 200
X = np.random.rand(n_samples, 2) * 100  # ki boyutlu veri (rnein gelir ve harcama skoru)

# K-means modelini eitelim
kmeans = KMeans(n_clusters=5, init='k-means++', max_iter=300, n_init=10, random_state=0)
y_kmeans = kmeans.fit_predict(X)

# Kmeleme sonularn grselletirelim
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], s=100, c='red', label='Kme 1')
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], s=100, c='blue', label='Kme 2')
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1], s=100, c='green', label='Kme 3')
plt.scatter(X[y_kmeans == 3, 0], X[y_kmeans == 3, 1], s=100, c='cyan', label='Kme 4')
plt.scatter(X[y_kmeans == 4, 0], X[y_kmeans == 4, 1], s=100, c='magenta', label='Kme 5')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='yellow', label='Merkezler')
plt.title('Rastgele Veri ile Mteri Segmentasyonu')
plt.xlabel('zellik 1')
plt.ylabel('zellik 2')
plt.legend()
plt.show()
```
